```python
# Importando bibliotecas

import pandas as pd

import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import plotly.express as px
import yfinance as yf
from selenium.webdriver import Firefox, FirefoxOptions
from selenium.webdriver.common.by import By
import pandas as pd
import time

```

```python

url = 'https://hedgefollow.com/funds/Berkshire+Hathaway'

options = FirefoxOptions()
# options.headless = True
options.add_argument('-headless')


driver = Firefox(options=options)
driver.get(url)

time.sleep(3)

table = driver.find_element(By.ID, 'dgtopHolders')

headers = []
rows = []
#print('--- headers ---')
header_row = table.find_elements(By.TAG_NAME, 'tr')[0]
for cell in header_row.find_elements(By.TAG_NAME, 'th'):
    headers.append(cell.text)

# print(headers)

# print('--- data ---')
data_rows = table.find_elements(By.TAG_NAME, 'tr')[1:]
for row in data_rows:
    row_data = {}
    for i, cell in enumerate(row.find_elements(By.TAG_NAME, 'td')):
        row_data[headers[i]] = cell.text
    rows.append(row_data)
    #print(row_data)

driver.quit()

df_bh = pd.DataFrame(rows)


df_bh.head()
```

```python
# Eliminando las columnas innecesarias
df_bh = df_bh.drop(['Ownership History', 'Average Buy Price', 'Price History'], axis=1)
```
